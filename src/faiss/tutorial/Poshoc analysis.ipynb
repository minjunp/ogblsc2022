```python
import numpy as np
import pandas as pd
from ogb.lsc import WikiKG90Mv2Dataset
import sys
import os
```

```python
rootdir = '/db2/users/minjunpark/ogb/rawdata'
dataset = WikiKG90Mv2Dataset(root=str(rootdir))
```

```python
# Validation dataset
valid_task = dataset.valid_dict['h,r->t'] # get a dictionary storing the h,r->t task.
hr = valid_task['hr']
t = valid_task['t']
```

```python
def check_candidates(validfile):
    basepath = '/db2/users/minjunpark/ogb/faiss/processed_data'
    valid_candidates = np.load(f'{basepath}/{validfile}')

    assert len(valid_candidates) == len(t)

    print(len(valid_candidates))
    valid_candidates
    
    num_included = 0
    for i in range(len(t)):
        if t[i] in valid_candidates[i]:
            num_included += 1
            
    print(f'Total cases: {len(t)}')
    print(f'Total number included: {num_included}')
    print(f'Success probability: {round(num_included/len(t), 4)}')
    
    return valid_candidates
```

```python
validfile = 'I_valid_candidates.npy'
check_candidates(validfile)
```
